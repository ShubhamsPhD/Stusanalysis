# Comparison With SPP

Seeing if we can reproduce Joke's neutron scattering length density profiles from 2 of her papers.

First, define our function for calculating the scattering length density given a set of structure factors.

```python
import numpy as np

def rho(F_hs, d, z_range=(-30, 30)):
    z = np.linspace(z_range[0], z_range[1], 501)
    rho_z = np.zeros_like(z)
    for i, F_h in enumerate(F_hs):
        rho_z += F_h * np.cos(2 * np.pi * (i+1) * z / d)
    rho_z *= 2 / d
    return np.vstack((z, rho_z)).T
```

In their first paper, they localize ceramide in SPP by deuterating the FA tail of CER NS and comparing the difference profiles of the protonated and deuterated samples. Here, we compare our density profile of the uCER2 FA tail from the simulation to the ones they back out of the scattering experiments.

```python
import matplotlib.pyplot as plt
import timplotlib as tpl
z_range = (-53.6/2, 53.6/2)
protonated = rho([-41.23, 19.67, -13.99, 14.17, -5.19], 53.6)#, z_range)
deuterated = rho([-43.68, 26.04, -13.99, 18.03, -6.10], 53.6, z_range)
deuterated[:, 1] += 0.55
for profile, label in zip([protonated, deuterated], ['Protonated', 'Deuterated']):
    pass
    #ax.plot(profile[:, 0], profile[:, 1], label=label)
difference = deuterated[:, 1] - protonated[:, 1]
protonated[:, 1] -= np.amin(protonated[:, 1])
rcfile = '/Users/tcmoore3/papers/my_papers/cer2-chol-atomistic/figs/matplotlibrc'
with plt.rc_context(fname=rcfile):
    fig, ax = plt.subplots(figsize=(3.25, 2.75))
    ax.plot(protonated[:, 0], protonated[:, 1], label='Experiment')
    data = np.loadtxt('1-3-4-4/equil2/end-data/profile_neutron-scattering.txt')
    data[:, 1] -= np.amin(data[20:-20, 1])
    #data[:, 1] += np.loadtxt('1-3-4-4/equil2/end-data/profile_bottom-ucer2-fa-tail-full.txt')[:, 1]
    trim = 1
    scale = np.amax(protonated[:, 1]) / np.amax(data[:, 1])
    ax.plot(data[trim:-trim, 0], data[trim:-trim, 1]*scale, label='Simulation')
    ax.set_xlim(-30, 30)
    ax.set_xlabel('$\mathbf{z}$, \u00c5')
    ax.set_ylabel('$\mathbf{NSLD}$, a.u.')
    ax.set_xticks([-30, -20, -10, 0, 10, 20, 30])
    tpl.timize(ax)
    fig.tight_layout()
    fig.savefig('figures/nsld-comparison.pdf')
    plt.show()
```

So we see some nice overlap there. First notice that the profile from experiment has a lot more features; a lot of that is from fitting 5 cosine curves to the data. The peak in the middle of the profile from experiment is somewhat reproduced in the simulation data, at least in location. Also interesting is the fact that the minimum at ±18 A is where our profile starts to decrease near the headgroups. Let's see how things look when we compare the locations of CHOL and FFA between experiment and simulation.

```python
import numpy as np
import matplotlib.pyplot as plt
import timplotlib as tpl
from collections import OrderedDict
from matplotlib.ticker import FormatStrFormatter
import string
z_range = (-54.3/2, 54.3/2)
cer_z_range = (-20, 20)
Fh_s = OrderedDict([
    ('cerfa_deut', {'d': 53.6, 'F': [-43.68, 26.04, -13.99, 18.03, -6.10]}),
    ('ctail_deut', {'d': 54.1, 'F': [-26.3, 21.9, -22.3, 4.6, -14.6]}),
    ('chead_deut', {'d': 53.8, 'F': [-33.1, 22.1, -10.6, 5.4, -2.9]}),
    ('protonated', {'d': 54.3, 'F': [-26.8, 19.2, -12.3, 10.2, -5.1]}),
    ('cer-prot', {'d': 53.6, 'F': [-41.23, 19.67, -13.99, 14.17, -5.19]})])
#    ('ffa_deut'  , {'d': 54.1, 'F': [-36.9, 26.1, -8.4, 13.9, -4.6]})])
protonated_data = rho(Fh_s['protonated']['F'], Fh_s['protonated']['d'], z_range)
cer_protonated_data = rho(Fh_s['cer-prot']['F'], Fh_s['cer-prot']['d'], cer_z_range)
trans = {'chead_deut': 'chol-headgroups', 'ctail_deut': 'chol-tail', 'ffa_deut':'ffa',
         'cerfa_deut': 'ucer2-fa-tail-full'}
skip = 5
offset = -0.2
rcfile = '/Users/tcmoore3/papers/my_papers/cer2-chol-atomistic/figs/matplotlibrc'
with plt.rc_context(fname=rcfile):
    fig, axs = plt.subplots(nrows=3, figsize=(3.25, 6), sharex=True, sharey=False)
    for (deut, info), ax, letter in zip(Fh_s.items(), axs, string.ascii_lowercase):
        if deut == 'cerfa_deut':
            ref_data = cer_protonated_data
            multiplier = 1.0 #sum(Fh_s['cer-prot']['F']) / sum(info['F'])  # set sums equal
            data = rho(np.array(info['F'])*multiplier, info['d'], cer_z_range)
        else:
            ref_data = protonated_data
            multiplier = sum(Fh_s['protonated']['F']) / sum(info['F'])  # set sums equal
            data = rho(np.array(info['F'])*multiplier, info['d'], z_range)
        try:
            sim_data = np.loadtxt('1-3-4-4/equil2/end-data/profile_' + trans[deut] + '.txt')
        except FileNotFoundError:
            sim_data = np.loadtxt('1-3-4-4/equil2/end-data/profile_top-' + trans[deut] + '.txt')
            sim_data[:, 1] += np.loadtxt('1-3-4-4/equil2/end-data/profile_bottom-' + trans[deut] + '.txt')[:, 1]
        #data[:, 1] -= np.amin(data[:, 1])
        #ref_data[:, 1] -= np.amin(ref_data[:, 1])
        diff = data[:, 1] - ref_data[:, 1]
        #ax.plot(data[:, 0], data[:, 1], label='deut')
        #ax.plot(ref_data[:, 0], ref_data[:, 1], label='prot')
        sim_data[:, 1] *= (np.max(diff) - np.min(diff)) / (np.max(sim_data[:, 1]) - np.min(sim_data[:, 1]))
        ax.plot(data[:, 0], diff-np.amin(diff),
                label='Experiment')
        ax.plot(sim_data[skip:-skip, 0], sim_data[skip:-skip, 1], label='Simulation')
        ax.yaxis.set_label_coords(offset, 0.5)
        ax.set_xlim(-28, 28)
        ax.set_ylabel(u'$\mathbf{\\rho(z)}$, a.u.')
        ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
        ax.text(0.05, 0.95, '{0})'.format(letter),
               transform=ax.transAxes, ha='left', va='top', size=12,
               bbox={'facecolor': 'w', 'alpha': 0.7, 'ec': 'none'})
        tpl.timize(ax, legend=(ax is axs[0]), legend_loc='upper right')
    axs[-1].set_xlabel(u'$\mathbf{z}$, \u00c5')
    fig.tight_layout()
    fig.savefig('figures/density-profile-compare.pdf')
    plt.show()
```
